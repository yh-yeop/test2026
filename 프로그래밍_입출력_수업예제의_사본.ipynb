문자열 Hello, Python, World를
Hello, Python, World
Python, World, Hello
으로 출력하는 프로그램을 작성해보자.

```python
s1='Hello, Python, World'
s2=s1.split()
print(f'{s2[0]} {s2[1]} {s2[2]} {s2[1]} {s2[2]}, {s2[0][:-1]}')
```

문자열 “Hello”, “Python”, “World”를 리스트에 넣은 후, “World”, “Python”, “Hello” 순서로 한 줄로 출력하는 프로그램을 작성해 보자.

```python
s3=[s.replace(',','') for s in s2]
print(*s3[::-1])
```

두 개의 단어를 빈칸으로 두고 한 줄로 입력받아 변수에 저장한 후, 순서를 바꾸어 한 줄에 하나씩 출력하는 프로그램을 작성해 보자.

```python
print(*input().split()[::-1],sep='\n')
# s1,s2=input().split()
# print(s2)
# print(s1)
```

세 개의 단어를 빈칸으로 두고 한 줄로 입력받아 변수에 저장한 후, 한 번에 하나씩 순서대로 출력하는 프로그램(리스트를 사용하세요)

```python
print(*input().split(),sep='\n')
```

다섯 개의 정숫값을 빈칸으로 두고 한 줄로 입력받아 리스트 변수에 저장한 후, 순서를 바꾸어 스페이스를 사이에 두고 한 줄로 출력하는 프로그램을 작성해 보자.

```python
print(*tuple(map(int,input().split()))[::-1],sep=' ')
```

일.월.년 형식으로 입력받은 값을 연.월.일 형식으로 바꾸어 출력하는 프로그램을 작성해 보자.

```python
print(*input().split()[::-1])
```

쉼표(,)를 사이에 둔 형식의 다섯 개의 정숫값들을 한 줄로 입력받아, 스페이스를 사이에 두고 한 줄로 출력하는 프로그램을 작성해 보자.

```python
print(*map(int,input().split(',')))
```
